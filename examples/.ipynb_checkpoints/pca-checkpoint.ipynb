### T-SNE NLF

```python
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns


fig, axs = plt.subplots(2, 2, figsize=(15, 15))
cofactors = ['NAD', 'NADP', 'FAD', 'S-Adenosyl-L-methionine']

for i, cofactor in enumerate(cofactors):
    ax = axs[i // 2, i % 2]

    df_without_cofactor = df_nlf.drop(columns=cofactors)
    df = StandardScaler().fit_transform(df_without_cofactor)
    df = pd.concat([pd.DataFrame(df, columns=df_without_cofactor.columns), df_nlf[cofactor]], axis=1)

    tsne = TSNE(n_components=2, n_iter=1000, random_state=42)
    tsne_results = tsne.fit_transform(df)

    sns.scatterplot(
        x=tsne_results[:,0], y=tsne_results[:,1],
        hue=cofactor,
        palette=sns.color_palette("hls", 3),
        data=df,
        legend="full",
        ax=ax
    ).set(title=f"t-SNE for {cofactor}")

plt.tight_layout()
plt.show()
plt.savefig('../data/encoding/tsne_all_cofactors_nlf.png')
plt.close()
```

```python
import pandas as pd
from propythia.feature_selection import FeatureSelection
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import ExtraTreesClassifier


def run_feature_selection(df, cofactors):
    df_all_cofactors = pd.DataFrame()

    for cofactor in cofactors:
        x_original = df.drop(['ProteinID', cofactor], axis=1).values
        target = df[cofactor].values
        columns_names = df.drop(['ProteinID', cofactor], axis=1).columns.tolist()

        feature_selection = FeatureSelection(x_original=x_original, target=target, columns_names=columns_names)

        select_model, x_transf_model, column_selected, feat_impo, feat_impo_df = feature_selection.run_from_model()

        df_selected = df.iloc[:,column_selected]
        df_selected[cofactor] = target

        df_all_cofactors = pd.concat([df_all_cofactors, df_selected], axis=1)

   
    df_all_cofactors = df_all_cofactors.loc[:,~df_all_cofactors.columns.duplicated()]
    
    for cofactor in cofactors:
        if cofactor in df_all_cofactors.columns:
            temp = df_all_cofactors.pop(cofactor)  # remove column and store it in temp
            df_all_cofactors[cofactor] = temp

    # Save the selected features to a TSV file
    df_all_cofactors.to_csv('../data/descriptores/linearSVC.tsv', sep='\t', index=False)

# Usage:
cofactors = ['FAD', 'NAD', 'FAD', 'NADP', 'S-Adenosyl-L-methionine']
run_feature_selection(df_descriptores, cofactors)
```

# ExtraTreesClassifier

```python
import pandas as pd
from propythia.feature_selection import FeatureSelection
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import ExtraTreesClassifier


def run_feature_selection(df, cofactors):
    df_all_cofactors = pd.DataFrame()

    for cofactor in cofactors:
        x_original = df.drop(['ProteinID', cofactor], axis=1).values
        target = df[cofactor].values
        columns_names = df.drop(['ProteinID', cofactor], axis=1).columns.tolist()

        feature_selection = FeatureSelection(x_original=x_original, target=target, columns_names=columns_names)

        select_model, x_transf_model, column_selected, feat_impo, feat_impo_df = feature_selection.run_from_model(ExtraTreesClassifier(n_estimators=50))

        df_selected = df.iloc[:,column_selected]
        df_selected[cofactor] = target

        df_all_cofactors = pd.concat([df_all_cofactors, df_selected], axis=1)


    df_all_cofactors = df_all_cofactors.loc[:,~df_all_cofactors.columns.duplicated()]
    
    for cofactor in cofactors:
        if cofactor in df_all_cofactors.columns:
            temp = df_all_cofactors.pop(cofactor)  
            df_all_cofactors[cofactor] = temp


    df_all_cofactors.to_csv('../data/descriptores/trees.tsv', sep='\t', index=False)


cofactors = ['FAD', 'NAD', 'FAD', 'NADP', 'S-Adenosyl-L-methionine']
run_feature_selection(df_descriptores, cofactors)

```

```python
import pandas as pd
from propythia.feature_selection import FeatureSelection
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import ExtraTreesClassifier



def run_feature_selection(df, cofactors):
    df_all_cofactors = pd.DataFrame()

    for cofactor in cofactors:
        x_original = df.drop(['ProteinID', cofactor], axis=1).values
        target = df[cofactor].values
        columns_names = df.drop(['ProteinID', cofactor], axis=1).columns.tolist()

        feature_selection = FeatureSelection(x_original=x_original, target=target, columns_names=columns_names)

        select_model, x_transf_model, column_selected, feat_impo, feat_impo_df = feature_selection.run_from_model(LogisticRegression(C=0.1, penalty="l2", dual=False))

        df_selected = df.iloc[:,column_selected]
        df_selected[cofactor] = target

        df_all_cofactors = pd.concat([df_all_cofactors, df_selected], axis=1)

    # Remove duplicate columns
    df_all_cofactors = df_all_cofactors.loc[:,~df_all_cofactors.columns.duplicated()]
    for cofactor in cofactors:
        if cofactor in df_all_cofactors.columns:
            temp = df_all_cofactors.pop(cofactor)  # remove column and store it in temp
            df_all_cofactors[cofactor] = temp

    # Save the selected features to a TSV file
    df_all_cofactors.to_csv('../data/descriptores/logistic.tsv', sep='\t', index=False)

# Usage:
cofactors = ['FAD', 'NAD', 'FAD', 'NADP', 'S-Adenosyl-L-methionine']
run_feature_selection(df_descriptores, cofactors)

```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


def plot_pca(df, cofactor):
    X = df.drop(columns=['NAD', 'NADP', 'FAD','S-Adenosyl-L-methionine']).values
    y = df[cofactor].values

    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(X)

    pca = PCA(n_components=2)  
    pca_result = pca.fit_transform(scaled_features)
    print(pca_result)

    # Separate points based on the value of the cofactor (0 or 1)
    cofactor_0_indices = np.where(y == 0)
    cofactor_1_indices = np.where(y == 1)

    plt.figure(figsize=(8, 6))

    # Plot points with cofactor value 0
    plt.scatter(pca_result[cofactor_0_indices, 0], pca_result[cofactor_0_indices, 1], color='blue', label=f'{cofactor}=0')

    # Plot points with cofactor value 1
    plt.scatter(pca_result[cofactor_1_indices, 0], pca_result[cofactor_1_indices, 1], color='red', label=f'{cofactor}=1')

    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.title(f'PCA of selected features + {cofactor}')
    plt.legend()
    plt.show()

# Usage:
cofactors = ['FAD', 'NAD', 'NADP', 'S-Adenosyl-L-methionine']
for cofactor in cofactors:
    plot_pca(df_feat1, cofactor)
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns


def plot_kmeans(df, cofactor):
    X = df.drop(columns=['NAD', 'NADP', 'FAD', 'S-Adenosyl-L-methionine']).values
    y = df[cofactor].values

    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(X)

    k = 2
    kmeans = KMeans(n_clusters=k, max_iter=1000, n_init=10)
    kmeans.fit(scaled_features)

    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_

    for i in range(k):
        ds = scaled_features[np.where(labels==i)]

        plt.plot(ds[:,1],ds[:,0],'o', label="cluster " + str(i))  
        plt.legend()
        # plot the centroids
        lines = plt.plot(centroids[i,0],centroids[i,1],'kx')
        # make the centroid x's bigger
        plt.setp(lines,ms=10.0)    # x size
        plt.setp(lines,mew=2.0)    # line width
    plt.show()

# Usage:
cofactors = ['FAD', 'NAD', 'NADP', 'S-Adenosyl-L-methionine']
for cofactor in cofactors:
    plot_kmeans(df_feat1, cofactor)
```

```python
from sklearn.manifold import MDS

def plot_mds(df, cofactor):
    X = df.drop(columns=['NAD', 'NADP', 'FAD', 'S-Adenosyl-L-methionine']).values
    y = df[cofactor].values

    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(X)

    df_scaled = pd.concat([pd.DataFrame(scaled_features), pd.DataFrame(y)], axis=1)
    df_scaled.columns = list(df.drop(columns=['NAD', 'NADP', 'FAD', 'S-Adenosyl-L-methionine']).columns) + [cofactor]

    mds = MDS(n_components=2, normalized_stress="auto")
    X_mds = mds.fit_transform(scaled_features)

    for specie in df_scaled[cofactor].unique():
        sp = df_scaled.index[df_scaled[cofactor]==specie]-1
        plt.plot(X_mds[sp,0], X_mds[sp,1], 'o' , label=specie)
    plt.title("MDS")
    plt.legend(loc='best', shadow=False)
    plt.show()

# Usage:
cofactors = ['FAD', 'NAD', 'NADP', 'S-Adenosyl-L-methionine']
for cofactor in cofactors:
    plot_mds(df_feat1, cofactor)
```

```python
import pandas as pd
from propythia.feature_selection import FeatureSelection
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import ExtraTreesClassifier


def run_feature_selection(df, cofactors):
    df_all_cofactors = pd.DataFrame()

    for cofactor in cofactors:
        x_original = df.drop(['ProteinID', cofactor], axis=1).values
        target = df[cofactor].values
        columns_names = df.drop(['ProteinID', cofactor], axis=1).columns.tolist()

        feature_selection = FeatureSelection(x_original=x_original, target=target, columns_names=columns_names)

        select_model, x_transf_model, column_selected, feat_impo, feat_impo_df = feature_selection.run_from_model(ExtraTreesClassifier(n_estimators=50))

        df_selected = df.iloc[:,column_selected]
        df_selected[cofactor] = target

        df_all_cofactors = pd.concat([df_all_cofactors, df_selected], axis=1)


    df_all_cofactors = df_all_cofactors.loc[:,~df_all_cofactors.columns.duplicated()]
    
    for cofactor in cofactors:
        if cofactor in df_all_cofactors.columns:
            temp = df_all_cofactors.pop(cofactor)  
            df_all_cofactors[cofactor] = temp


    df_all_cofactors.to_csv('../data/descriptores/trees.tsv', sep='\t', index=False)


cofactors = ['FAD', 'NAD', 'FAD', 'NADP', 'S-Adenosyl-L-methionine']
run_feature_selection(df_descriptores, cofactors)

```
