### Heatmap of Cofactor Co-occurrence
heatmap can visualize the co-occurrence of different cofactors within the same proteins, which is useful to understand relationships between cofactors.

```python
cofactor_columns = df.columns[1:-1]  # Exclude 'ProteinID' and 'Sequence' columns
co_occurrence_matrix = df[cofactor_columns].T.dot(df[cofactor_columns])

plt.figure(figsize=(10,8))
sns.heatmap(co_occurrence_matrix, annot=True, cmap="coolwarm", fmt="d")
plt.title('Co-occurrence of Cofactors Across Proteins')
plt.xlabel('Cofactor')
plt.ylabel('Cofactor')
plt.show()
```

### Pairwise Cofactor Analysis
Analyze how frequently each pair of cofactors appears together in the same protein. This can be done using a simple pairwise comparison

```python
from itertools import combinations

cofactor_columns = df.columns[1:-1]
pairwise_counts = {}

for cofactor1, cofactor2 in combinations(cofactor_columns, 2):
    pairwise_counts[f"{cofactor1} & {cofactor2}"] = df[(df[cofactor1] == 1) & (df[cofactor2] == 1)].shape[0]

pairwise_counts = pd.Series(pairwise_counts).sort_values(ascending=False)
print(pairwise_counts.head(10))  # Display the top 10 most common cofactor pairs
```

```python
def multilabel_train_test_split(X, y, stratify=None, test_size=0.2, train_size=0.65, val_size = 0.10, random_state=42):
    X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=test_size, stratify=stratify, random_state=random_state)
    X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=val_size, stratify=stratify, random_state=random_state)
    return X_train, X_val, X_test, y_train, y_val, y_test
```
