### Heatmap of Cofactor Co-occurrence
heatmap can visualize the co-occurrence of different cofactors within the same proteins, which is useful to understand relationships between cofactors.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df1' is the dataset with cofactor columns
cofactor_columns = df1.columns[1:-1]  # Exclude 'ProteinID' and 'Sequence' columns

# Calculate co-occurrence matrix
co_occurrence_matrix = df1[cofactor_columns].T.dot(df1[cofactor_columns])

# Ensure the diagonal has the total count for each cofactor
for cofactor in cofactor_columns:
    co_occurrence_matrix.loc[cofactor, cofactor] = df1[cofactor].sum()

# Set color scale based on the range of values
vmin = 0  # Minimum value
vmax = co_occurrence_matrix.max().max()  # Maximum value in the matrix

# Plot the heatmap with adjusted color scale
plt.figure(figsize=(10,8))
sns.heatmap(co_occurrence_matrix, annot=True, cmap="YlOrRd", fmt="d", vmin=vmin, vmax=vmax)
plt.title('Co-occurrence of Cofactors Across Proteins')
plt.xlabel('Cofactor')
plt.ylabel('Cofactor')
plt.show()
```

### Pairwise Cofactor Analysis
Analyze how frequently each pair of cofactors appears together in the same protein. This can be done using a simple pairwise comparison

```python
from itertools import combinations

cofactor_columns = df.columns[1:-1]
pairwise_counts = {}

for cofactor1, cofactor2 in combinations(cofactor_columns, 2):
    pairwise_counts[f"{cofactor1} & {cofactor2}"] = df[(df[cofactor1] == 1) & (df[cofactor2] == 1)].shape[0]

pairwise_counts = pd.Series(pairwise_counts).sort_values(ascending=False)
print(pairwise_counts.head(10))  # Display the top 10 most common cofactor pairs
```
